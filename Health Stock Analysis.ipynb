### Function to Plot all Parameters of a Stock 

```python
def plot_all(stock_name,name):
    stock_name[['Open','Close','Adj Close']].plot(legend=True, figsize=(15,6))
    plt.title('Parameters for %s' %name)
```

```python


plot_all(UNH,"UNH")
plot_all(CVS,"CVS")
plot_all(JNJ, "JNJ")
plot_all(AMGN,"AMGN")
plot_all(GILD,"GILD")
```

### Calculating Moving Average

```python
def mov(stock):
    movav= [10,20,50]
    for ma in movav:
        name=("MA of Adj Close for %s"%(str(ma)))
        stock[name]= pd.rolling_mean(stock['Adj Close'],ma)
    return(stock)
```

```python
mov(UNH)
mov(CVS)
mov(JNJ)
mov(AMGN)
mov(GILD)
```

## Function to plot MA values

```python
def plotma(stock,name):
    stock[['Adj Close', 'MA of Adj Close for 10','MA of Adj Close for 20','MA of Adj Close for 50']].plot(subplots=False,
                                                                                                         figsize=(15,6))
    plt.title('Moving Average Graph for %s' %name)
```

```python
plotma(JNJ,'JNJ')
```

### Risk Analysis

#### Here the Risk Analysis is performed by Quantified by plotting the Std of the Daily Return with the Expected Return values

```python
adjclose= adj_close_pc.dropna()
```

```python
plt.xlim([-0.001,0.0008])
plt.ylim([0.01,0.025])
plt.scatter(adjclose.mean(), adjclose.std())
plt.xlabel('Expected Returns(Mean)')
plt.ylabel('Standard Deviation ')
for label, x, y in zip(adjclose.columns, adjclose.mean(), adjclose.std()):
    plt.annotate(
        label, 
        xy = (x, y), xytext = (50, 50),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        arrowprops = dict(arrowstyle = '-', connectionstyle = 'arc3,rad=-0.3'))
```

### It looks like AMGN and UNH are the two stocks which promises more return with a lower standard deviation
